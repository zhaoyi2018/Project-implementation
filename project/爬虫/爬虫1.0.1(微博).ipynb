```
def re_filter(content):
    '''
    使用re工具, 处理冗余信息
    '''
    import re

    pattern1 = re.compile(r'#[^#]+#')
    pattern2 = re.compile(r"[a-zA-Z]")
    return re.sub(pattern2, "", re.sub(pattern1, "", content))
```

```
def get_info(res, get_content=True):
    '''
    获取指定的内容(题目和评论)或问题id
    '''
    from bs4 import BeautifulSoup

    if res.status_code != 200:
        print("ip被封!, 等候20s")
        sleep(20)
        if get_content:
            return 0, 0
        else:
            return []
    contents_dict = res.json()["data"]
    if get_content:
        contents_list = []
        # 获取下一页
        new_page = contents_dict["cardlistInfo"]["page"]
        # 获取主题
        title_name = contents_dict["cardlistInfo"]["cardlist_title"]
        for item in contents_dict["cards"]:
            # 判断评论类型
            if item["card_type"] != 9:
                continue
            # 如果是, 就获取评论内容和题目
            content = ""
            html = item["mblog"]["text"]
            bf = BeautifulSoup(html)
            for p in bf.stripped_strings:
                content += p
            # 使用re工具, 将##内容全部清除
            content = re_filter(content)
            contents_list.append((title_name, content))
        return new_page, contents_list
    else:
        keyword_list = []
        for item in contents_dict["cards"][0]["group"]:
            content = ""
            keyword = str(item["title_sub"])
            keyword_list.append(keyword)
        return keyword_list
```

```
def web_grap(keyword):
    '''
    网页抓取
    '''
    import requests
    import pandas as pd
    
    base_url = "https://m.weibo.cn/api/container/getIndex?containerid=100103type%3D1%26q%3D{}&page_type=searchall&page={}"
    # headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.183 Safari/537.36 Edg/86.0.622.63"}
    page = 1
    print("\n抓取热度:{}".format(keyword))
    while True:
        url = base_url.format(keyword, page)
        random_sleep()
        res = requests.get(url.format(keyword, page), headers=random_header())
        new_page, contents_list = get_info(res)

        # 判断ip是否被封
        if new_page==0:
            print("再次申请...")
            continue

        ## 存储数据到csv中
        dataset = pd.DataFrame(contents_list, columns=["title", "comment"])
        dataset.to_csv("data.csv", mode="a", header=False, index=False, sep="\t")
        print("第{}页抓取成功!".format(page))

        ## 抓取下一页
        if new_page==page or new_page is None:
            print("{}抓取结束!".format(keyword))
            break
        else:
            page = new_page
```

```
def to_more_grap():
    '''
    用于获取相关热度的关键词
    '''
    import requests
    
    # 准备网站
    base_url = "https://m.weibo.cn/api/container/getIndex?containerid=231583&page_type=searchall"
    
    ## 准备https头部
    # headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.183 Safari/537.36 Edg/86.0.622.63"}
    # ## 睡眠
    random_sleep()
    ## 开始访问
    res = requests.get(base_url, headers=random_header())
    ## 处理json数据, 获取更多问题号
    contents_list = get_info(res, get_content=False)
    return contents_list
```

```
def set_grap(question_num=1):
    '''
    集成爬虫
    '''
    # 爬取关键词
    keyword_list = to_more_grap()
    web_grap(keyword_list[0])
    question_num -= 1

    i = 1
    while question_num:
        temp_i = i
        while len(keyword_list) == i:
            keyword_list += to_more_grap()
            keyword_list = sorted(set(keyword_list),key=keyword_list.index)
            temp_i -= 1
            if temp_i == -1:
                print("因热度关键词一致, 故抓取结束!")
                return
        web_grap(keyword_list[i])
        i += 1
        question_num -= 1
    print("抓取任务结束!")
```
